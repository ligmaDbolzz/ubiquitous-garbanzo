```python
with open('RYT.txt') as f:
    passage = f.read()
passage
```

#Tokenizing

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sent_tokenize(passage)
```

```python
words = word_tokenize(passage)
words
```

#Stemming

```python
from nltk.stem import PorterStemmer
import pandas as pd
```

```python
stemmed_words = [PorterStemmer().stem(word) for word in words]
stemmed_words
```

#Lemmatizing


```python
from nltk.stem import WordNetLemmatizer
```

```python
wnl = WordNetLemmatizer()
lemmatized_words = [wnl.lemmatize(word).lower() for word in words]
lemmatized_words
```

```python
comp = pd.DataFrame({'Before': words,
     'Stemming': stemmed_words,
     'Lemmatizing': lemmatized_words
    })
comp
```
